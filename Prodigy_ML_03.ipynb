```python
import os

# Correct the folder path using raw string
folder_path = r"Users\elkha\Downloads\dogs-vs-cats"
os.makedirs(folder_path, exist_ok=True)

# Define path
confusion_image_path = os.path.join(folder_path, 'confusion matrix.png')
classification_file_path = os.path.join(folder_path, 'classification_report.txt')
model_file_path = os.path.join(folder_path, "svm_model.pkl")
```

```python
import os
import cv2
import numpy as np
from tqdm import tqdm

# Define the correct train directory path
train_dir = r"C:\Users\elkha\Downloads\dogs-vs-cats\train\train"
test_dir = r"C:\Users\elkha\Downloads\dogs-vs-cats\test1\test1"

# Ensure the train directory exists
if not os.path.exists(train_dir):
    print(f"Train directory does not exist: {train_dir}")
else:
    print(f"Train directory exists: {train_dir}")

# List the image files in the train directory
train_images = [f for f in os.listdir(train_dir) if os.path.isfile(os.path.join(train_dir, f))]
print(f"Number of training images found: {len(train_images)}")

features = []
labels = []
image_size = (32, 32)

# Process train images
for image in tqdm(train_images, desc="Processing Train Images"):
    print(f"Processing image: {image}")
    if image.lower().startswith('cat'):
        label = 0
    elif image.lower().startswith('dog'):
        label = 1
    else:
        print(f"Unknown label for image: {image}")
        continue
    
    image_path = os.path.join(train_dir, image)
    image_read = cv2.imread(image_path)
    
    if image_read is None:
        print(f"Failed to load image: {image_path}")
        continue
    
    image_resized = cv2.resize(image_read, image_size)
    image_normalized = image_resized / 255.0
    image_flatten = image_normalized.flatten()
    features.append(image_flatten)
    labels.append(label)

# Convert features and labels to numpy arrays for machine learning
features = np.array(features)
labels = np.array(labels)

print("Processing complete.")
```

```python
del train_images
```

```python
features = np.asarray(features)
labels = np.asarray(labels)

# train test split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, shuffle=True, random_state=42)
```

```python
del features
del labels

```

```python
# PCA, SVM, & Pipeline
n_components = 0.8
pca = PCA(n_components=n_components)
svm = SVC()
pca = PCA(n_components=n_components, random_state=42)
pipeline = Pipeline([
    ('pca', pca),
    ('svm', svm)
])
```

```python
param_grid = {
    'pca__n_components': [2, 1, 0.9],
    'svm__kernel': ['linear',  'sigmoid'],
}
```

```python

start_time = time.time() 

grid_search = GridSearchCV(pipeline, param_grid, cv=3, verbose=4)
grid_search.fit(X_train, y_train)


end_time = time.time()
```

```python
del X_train
del y_train
```

```python
best_pipeline = grid_search.best_estimator_
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Parameters: ", best_params)
print("Best Score: ", best_score)
```

```python
accuracy = best_pipeline.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python
y_pred = best_pipeline.predict(X_test)


target_names = ['Cat', 'Dog']
classification_rep = classification_report(y_test, y_pred, target_names=target_names)
print("Classification Report:\n", classification_rep)

with open(classification_file_path, 'w') as file:
    file.write(classification_rep)
```
